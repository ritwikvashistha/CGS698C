**Q2.  Consider the height-weight correlation example.  Build a Bayesian model for the correlation by finding posteriors for Î²0,Î²1,Ï as follows**

* First create a data generator by assuming reasonable values for the three parameters Î²0,Î²1,Ï. To generatea data points first draw a height valuehâ¼N(Î¼h,Ïh) (choose Î¼h,Ïh suitably), then use the regression equation to generate a predicted value, say  Ëw, draw the sample data point using wâ¼N( Ëw,Ï).

```python
import matplotlib.pyplot  as plt
import numpy as np
from scipy.stats import beta
from matplotlib.figure import Figure
import pymc3 as pm
import arviz as az
from scipy.stats import uniform
from numpy.random import normal
import pandas as pd
from random import sample
from statistics import mean
from statistics import stdev
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def data_gen(trials):
    beta_0=-310                    #TRUE VALUE
    beta_1=7                       #TRUE VALUE   
    sigma=10                       #TRUE VALUE 
    h=normal(65,7,trials)
    w=[]
    w1=[]
    for i in range(0,trials):
        w.append(normal(beta_0+beta_1*h[i],sigma,1))
        w1.append(w[i][0])
        
    return(h,w1)
```

* For two priors given below compute the posterior distributions for the three parameters for sample sizes of 10, 100 and 1000 and plot the distribution and the 95% HDI.The two priors are:  i) a non-informative prior (done in class) and ii) a more informed prior created bydrawing 10 data points using the generator in part (a) and then using N(Î¼â²,Ïâ²) as the prior where Î¼â²,Ïâ²are the sample mean and sample standard deviation of the ten-sized sample.

For Non Informative Prior

```python
def posterior_0(h,w):   
 with pm.Model() as model:
    alpha=pm.Uniform('alpha',-500,500)
    beta=pm.Uniform('beta',-10,10)
    std=pm.Uniform('sd',0,100)
    w_pred = pm.Normal('y_pred', mu= alpha + beta*h, sd=std, observed=w)
    trace=pm.sample(2000,cores=1)
    az.plot_posterior(trace, credible_interval=0.95)
```

FOR N=10

```python
h,w=data_gen(10)
posterior_0(h,w)
```

FOR N=100

```python
h,w=data_gen(100)
posterior_0(h,w)
```

FOR N=1000

```python
h,w=data_gen(1000)
posterior_0(h,w)
```

For Informative Prior

```python
def posterior_1(h,w,bo,b1,sigma):   
 with pm.Model() as model:
    alpha=pm.Normal('alpha',mu=bo,sd=150)
    beta=pm.Normal('beta',b1,150)
    std=pm.Normal('sd',sigma,150)
    w_pred = pm.Normal('y_pred', mu= alpha + beta*h, sd=std, observed=w)
    trace=pm.sample(4000,cores=1)
    az.plot_posterior(trace, credible_interval=0.95)
```

FOR N=100

```python
h,w=data_gen(100)
sample_h=np.array(sample(list(h),10)).reshape((-1, 1))
sample_w=sample(list(w),10)

model = LinearRegression().fit(sample_h, sample_w)
bo=model.intercept_
b1=model.coef_[0]
w_predicted = model.predict(sample_h)
rms = mean_squared_error(sample_w, w_predicted, squared=False)
posterior_1(h,w,bo,b1,rms)
```

FOR N=1000

```python
h,w=data_gen(1000)
sample_h=np.array(sample(list(h),10)).reshape((-1, 1))
sample_w=sample(list(w),10)

model = LinearRegression().fit(sample_h, sample_w)
bo=model.intercept_
b1=model.coef_[0]
w_predicted = model.predict(sample_h)
rms = mean_squared_error(sample_w, w_predicted, squared=False)
posterior_1(h,w,bo,b1,rms)
```
